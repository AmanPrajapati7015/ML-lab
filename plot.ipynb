```python
import numpy as np
import matplotlib.pyplot as plt

time = np.loadtxt('./oil.csv',  delimiter=',', usecols=1)
nitro = np.loadtxt('./oil.csv',  delimiter=',', usecols=5)

# print(time)
# print(nitro)

plt.scatter(nitro, time, marker='o')

# Add labels & title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Matplotlib Plot')

plt.show()

```
